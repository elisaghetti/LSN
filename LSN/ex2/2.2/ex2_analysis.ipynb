```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
```

## Lattice random walks

```python
file= 'RW_latt.csv'
df = pd.read_csv(file, sep = '\t', header=None, names = ['walk_id','x', 'y','z'])
df.head()
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
# Colori distinti per ogni random walk
id_values = df['walk_id'].unique()
num_walks_to_plot = 10
colors = plt.get_cmap("tab20", num_walks_to_plot)

for id in range(num_walks_to_plot):
    # Plotta il random walk
    ax.plot(df[df['walk_id'] == id]['x'], df[df['walk_id'] == id]['y'], df[df['walk_id'] == id]['z'], label=f"Walk {id+1}", color=colors(id))

# Etichette e legenda
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Primi 10 Random Walks")
ax.legend()

plt.show()
```

```python
file = "latt_distance.csv"

df= pd.read_csv(file,sep ='\t', header=None, names = ['step_id','dist','err'])
df.head()
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)

x_vals = np.linspace(min(df['step_id']), max(df['step_id']), 100)
y_vals = np.sqrt(x_vals)
ax.plot(x_vals, y_vals, label=r'$y=\sqrt{x}$', color='red', linestyle='--')

ax.errorbar(df['step_id'],df['dist'],yerr=df['err'])


plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
```

```python
N=10000;
N_blocks= 100; 
n= N/N_blocks;

x = np.arange(N_blocks) * n  
```

Integrale con metodo della media, p(x) =1

```python
file = "integral.csv"
df = pd.read_csv(file, sep='\t')
df.head()
```

perchè il primo errore è nullo? -> è giusto, vedi in fondo a notebook 1

```python
plt.figure()
type= "MM"
plt.errorbar(x, df[df['TYPE']== type]['VALUE'], yerr=df[df['TYPE']== type]['ERROR'], fmt='o',capsize=5, label="Data with Error Bars")
plt.title('Integral with importance sampling method')
plt.grid()

plt.hlines(1., 0, N, colors='r')
plt.show
```

Integrale con importance sampling: p(x)=2x (sviluppo integranda attorno a 1), g(x)= integranda/2x

```python
plt.figure()
type= "IS"
plt.errorbar(x, df[df['TYPE']== type]['VALUE'], yerr=df[df['TYPE']== type]['ERROR'], fmt='o', capsize=5, label="Data with Error Bars")
plt.title('Integral with importance sampling method')
plt.grid()

plt.hlines(1., 0, N, colors='r')
plt.show
```

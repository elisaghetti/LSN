```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
file = "SN_exp.csv"
df = pd.read_csv(file, sep ='\t',header=None, names = ['N','S_N'])
df.head()
```

```python
N_values = df['N'].unique()
colors = ['blue', 'red', 'green', 'orange']

plt.figure(figsize=(10, 6))
for i,val in enumerate(N_values):
    plt.hist(df[df['N'] == val]['S_N'], edgecolor = 'black', density=True, color=colors[i%len(colors)],label=f'N = {val}')
    plt.xlabel('Valori')
    plt.ylabel('Frequenza')
    plt.title('Esponenziale')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
file = "SN_cauchy.csv"
df = pd.read_csv(file, sep ='\t',header=None, names = ['N','S_N'])
df.head()
```

```python
N_values = df['N'].unique()
colors = ['blue', 'red', 'green', 'orange']

plt.figure(figsize=(10, 6))
for i,val in enumerate(N_values):
    plt.hist(df[df['N'] == val]['S_N'], bins=100,edgecolor = 'black', density=True, color=colors[i%len(colors)],label=f'N = {val}')
    plt.xlabel('Valori')
    plt.xlim(-2000,1000)
    plt.ylabel('Frequenza')
    plt.title('Cauchy-Lorentz')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

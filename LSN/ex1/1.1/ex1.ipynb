```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
M=10000             # Total number of throws
N=100                 # Number of blocks
L=int(M/N)  

x = np.arange(N) *L
```

```python
file = "statistics.csv"
df = pd.read_csv(file, sep = '\t')
df.head()
```

```python
# Plot media

type = "AVE";
plt.figure()

plt.errorbar(x, df[df['TYPE']== type]['VALUE'] - 0.5, yerr=df[df['TYPE']== type]['ERROR'], fmt='o', capsize=5, label="Data with Error Bars")

# Labels and grid
plt.xlabel('#throws')
plt.ylabel('<r> - 1/2')
plt.grid(True)
plt.legend()
```

```python
type = "VAR";
plt.figure()

plt.errorbar(x, df[df['TYPE']== type]['VALUE'] - 1./12., yerr=df[df['TYPE']== type]['ERROR'], fmt='o', capsize=5, label="Data with Error Bars")

# Labels and grid
plt.xlabel('#throws')
plt.ylabel('<r> - 1/2')
plt.grid(True)
plt.legend()
```

```python
with open("Chi.txt", 'r') as file:
    chi_squared = np.array([float(line.strip()) for line in file if line.strip()])

x_values = np.linspace(0, 1, len(chi_squared))
plt.figure()
plt.scatter(x_values,chi_squared)
plt.figure()
plt.hist(chi_squared, bins=30)
plt.xlabel('Chi-squared values')
plt.ylabel('Frequency')
plt.title('Chi-squared Histogram')
plt.grid(True)
plt.show()
```
